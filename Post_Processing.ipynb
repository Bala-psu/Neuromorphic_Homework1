```python
%matplotlib inline
import torch
import torch.nn as nn
from torch.nn.parameter import Parameter
import torch.nn.functional as F
import torch.optim as optim
import argparse
import os
import numpy as np
import time
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from enum import Enum
from lenet5 import lenet5
import math


def draw_linegraph(x, y1, y2, y3, color, figsize, legend, title, xlabel, ylabel):
    plt.figure(figsize = figsize)
    ax= plt.subplot()
    my_plot = plt.plot(x, y1, color = color, marker='s', linewidth = 7.0, markersize=9, label = 'LOSS (a=1)')
    plt.plot(x, y2, color = 'red', marker='s', linewidth = 7.0, markersize=9, label = 'LOSS (a=2)')
    plt.plot(x, y3, color = 'green', marker='s', linewidth = 7.0, markersize=9, label = 'LOSS (a=0.5)')
    plt.title(title, fontsize=30)
    plt.xlabel(xlabel, fontsize = 24)
    plt.xticks(ticks = x, fontsize = 20)
    plt.locator_params(axis='x', nbins= 25)
    plt.ylabel(ylabel, fontsize = 24)
    plt.yticks(fontsize = 20)
    plt.setp(ax.get_xticklabels(), rotation=30, ha='right')
    plt.grid()
    #plt.ylim(0, 110)
    plt.ticklabel_format(axis='y', scilimits=(-3,3), style='plain')
    plt.legend(prop={"size":30})
    return my_plot

def draw_line(x, y, color, figsize, legend, title, xlabel, ylabel):
    plt.figure(figsize = figsize)
    ax= plt.subplot()
    my_plot = plt.plot(x, y, color = color, marker='s', linewidth = 7.0, markersize=9)
    plt.title(title, fontsize=30)
    plt.xlabel(xlabel, fontsize = 24)
    plt.xticks(ticks = x, fontsize = 20)
    plt.locator_params(axis='x', nbins= 25)
    plt.ylabel(ylabel, fontsize = 24)
    plt.yticks(fontsize = 20)
    plt.setp(ax.get_xticklabels(), rotation=30, ha='right')
    plt.grid()
    #plt.ylim(0, 110)
    plt.ticklabel_format(axis='y', scilimits=(-3,3), style='plain')
    plt.legend()
    return my_plot


ACC = []
EPOCH = []
ITER = []
iter_count = 0
for epoch in range(1,21):
    RESULT = np.load('/home/mdl/bus44/Neuromorphic Computing/runs/Q1b/TRAIN_acc_'+ str(epoch) + '.npy')
    for iteration in range(len(RESULT)):
        if iteration % 100 == 0:
            ACC.append(RESULT[iteration])
            ITER.append(iter_count)
            iter_count = iter_count + 100
    EPOCH.append(epoch)
    

LOSS = []
for epoch in range(1,21):
    RESULT = np.load('/home/mdl/bus44/Neuromorphic Computing/runs/Q1b/TRAIN_loss_'+ str(epoch) + '.npy')
    for iteration in range(len(RESULT)):
        if iteration % 100 == 0:
            LOSS.append(RESULT[iteration])

LOSS_a2 = []
for epoch in range(1,21):
    RESULT = np.load('/home/mdl/bus44/Neuromorphic Computing/runs/Q1b/TRAIN_loss_a2sig_'+ str(epoch) + '.npy')
    for iteration in range(len(RESULT)):
        if iteration % 100 == 0:
            LOSS_a2.append(RESULT[iteration])            
            
LOSS_a05 = []
for epoch in range(1,21):
    RESULT = np.load('/home/mdl/bus44/Neuromorphic Computing/runs/Q1b/TRAIN_loss_a05sig_'+ str(epoch) + '.npy')
    for iteration in range(len(RESULT)):
        if iteration % 100 == 0:
            LOSS_a05.append(RESULT[iteration])  
            

TEST_RESULT = np.load('/home/mdl/bus44/Neuromorphic Computing/runs/Q1b/TEST_acc_20.npy', allow_pickle=True)

x = ITER
y1 = LOSS
y2 = LOSS_a2
y3 = LOSS_a05
color = 'blue'
legend = 'false'
figsize = (30,10)
title = "LOSS CURVE OF THE TWO LAYER FC NETWORK TRAINED ON MNIST"
xlabel = "TRAINING ITERATION"
ylabel = "LOSS"
draw_linegraph(x, y1,y2, y3, color, figsize, legend, title, xlabel, ylabel)

x = EPOCH
y = TEST_RESULT
color = 'darkgreen'
legend = 'false'
figsize = (30,10)
title = "TEST SET ACCURACY OF TWO LAYER FC NETWORK TRAINED ON MNIST"
xlabel = "EPOCH NUMBER"
ylabel = "ACCURACY (%)"
draw_line(x, y, color, figsize, legend, title, xlabel, ylabel)

#print(ACC)
#print(LOSS)
#print(TEST_RESULT)
```
